## 3. Lists

```python
a = [1, 2, 3]
b = list(a)
b[0] = 100

print(a)
print(b)
```

## 4. Conditionals (if / else)

```python
x = -3

if x > 0:
    print("Positive")
elif x < 0:
    print("Negative")
else:
    print("Zero")
```

```python
print("Big") if x > 10 else print("Small")
```

## 5. Loops (for / while)

```python
for x in "Oluwagbemiga":
    print(x)
```

```python
total = 0
for n in range(2, 101, 2):
    total += n
print(total)
```

```python
total = 1
value = 1

while total < 5000:
    value += 1
    total *= value

print(value, total)
```

## 6. Exercise 3.1

```python
weight = 80      # kg
height = 1.80    # m

bmi = weight / height**2

if bmi < 15:
    category = "Very severely underweight"
elif bmi < 16:
    category = "Severely underweight"
elif bmi < 18.5:
    category = "Underweight"
elif bmi < 25:
    category = "Normal"
elif bmi < 30:
    category = "Overweight"
elif bmi < 35:
    category = "Obese Class I"
elif bmi < 40:
    category = "Obese Class II"
else:
    category = "Obese Class III"

print("BMI:", round(bmi,2))
print("Category:", category)
```

## 10. Exercise 3.5

```python
metdata_orig = np.loadtxt("T_Alicante_EM.csv", delimiter=",", skiprows=1)
years = metdata_orig[:,0]
temp_data = metdata_orig[:,1:]


threshold = 30
counter = 0

for i in range(temp_data.shape[0]):
    if temp_data[i,:].max() >= threshold:
        print("Extreme year:", int(years[i]))
        counter += 1

print("Total extreme years:", counter)
```
