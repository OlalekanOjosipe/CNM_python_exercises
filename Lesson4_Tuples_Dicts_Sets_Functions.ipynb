## Tupules

```python
# Tuples: created with commas (parentheses optional)
tup_0 = 1, 2, 3, 4, 5
tup_1 = (2, 3, 4, 5, 6)
tup_2 = ((2, 3, -1), (0, 1, 2.4), 3, (-33, -22))

print("tup_0:", tup_0)
print("tup_1[3]:", tup_1[3])
print("tup_2[1]:", tup_2[1])

# Tuples are immutable (this would error if you uncomment it)
# tup_0[3] = 4

# But tuples can contain mutable objects (like lists) that can be modified
tup_3 = (4, (2, 3, -1), [4, 4, 5, 5], True)
tup_3[2].append(6)  # modifies the list inside the tuple
print("tup_3 after append:", tup_3)

# Joining and repeating tuples
print("tup_0 + tup_3:", tup_0 + tup_3)
print("4 * tup_0:", 4 * tup_0)

# Tuple unpacking
a, b, c, d = tup_2
print("unpacked:", a, b, c, d)

# Swapping values (classic tuple use)
print("before swap:", a, b)
a, b = b, a
print("after swap:", a, b)

# Tuple unpacking in loops
tup_loop = ((2, 3, -1), (4, 4, 5), (5, 6, 7), (5, -1, 0))
for var_1, var_2, var_3 in tup_loop:
    print(f"var_1={var_1}, var_2={var_2}, var_3={var_3}")
```

## Dictionaries

```python
# Creating dicts (hashes): key -> value
hash_0 = {
    "Guerras Médicas": ["Termópilas", "Artemisio", "Salamina", "Platea"],
    "Even integers": (0, 2, 4, 6, 8)
}

print("hash_0:", hash_0)
print("Battles:", hash_0["Guerras Médicas"])

# Modify a value (list) in place
hash_0["Guerras Médicas"].append("Micala")

# Add new key-value pair
hash_0["Fantastic Sea Creatures"] = ("Moby Dick", "The Kraken", "Mermaids")

print("updated hash_0:", hash_0)

# Create dict from sequence of pairs
seq = (1, 3), (2, 6), (3, 9), (4, 12)
dict_example = dict(seq)
print("dict_example:", dict_example)

# keys(), values(), items()
print("keys:", list(dict_example.keys()))
print("values:", list(dict_example.values()))
print("items:", list(dict_example.items()))
```

## Default values

```python
dict_example = {1: 3, 2: 6, 3: 9, 4: 12}

# get() avoids KeyError; can provide default
print("get(2):", dict_example.get(2))
print("get(99):", dict_example.get(99))               # returns None
print("get(99, default=0):", dict_example.get(99, 0)) # returns 0

# pop() removes and returns; can provide default
two_pop = dict_example.pop(2, "not found")
print("popped key 2:", two_pop)
print("after pop:", dict_example)

# delete with del
del dict_example[3]
print("after del key 3:", dict_example)

# ---- setdefault pattern: dict of lists ----
import numpy as np

random_nums = np.random.randint(0, 5000, 30)
last_digit_hash = {}

for number in random_nums:
    last_digit = number % 10
    last_digit_hash.setdefault(last_digit, []).append(number)

print("Grouped by last digit:", last_digit_hash)
```

## Sets

```python
# Sets contain unique elements (duplicates removed)
set_0 = {"a", 0, 1, "bc", 0.33, 0, 1}
set_1 = set(["a", "b", "c", "a", "a"])

print("set_0:", set_0)
print("set_1:", set_1)

# Union
print("union:", set_0.union(set_1))
print("union (|):", set_0 | set_1)

# Intersection
print("intersection:", set_0.intersection(set_1))
print("intersection (&):", set_0 & set_1)

# Difference
print("difference:", set_0.difference(set_1))
print("difference (-):", set_0 - set_1)

# Practical: remove duplicates
vals = [1, 1, 2, 3, 3, 3, 4]
print("unique:", set(vals))
```

## Comprehensions

```python
# List comprehension (with condition)
list_mults = [n for n in range(4000) if (n % 7 == 0 and n % 13 == 0)]
print("count:", len(list_mults))
print("first few:", list_mults[:10])

# Dict comprehension: number -> number^2 for even numbers < 10
sq_even = {n: n**2 for n in range(10) if n % 2 == 0}
print("dict comprehension:", sq_even)

# Set comprehension: unique remainders mod 7 from 0..49
mods = {n % 7 for n in range(50)}
print("set comprehension:", mods)

# sorted() with key
import numpy as np
random_nums = np.random.randint(0, 1000, 10)
print("random_nums:", random_nums)
print("sorted numeric:", sorted(random_nums))
print("sorted as strings:", sorted(random_nums, key=str))

# reversed()
for n in reversed(range(5)):
    print("reversed:", n)

# zip(): combine sequences
names = ["Lisa", "Auxi", "Julia", "Lisanna", "Curro"]
rand_a = np.random.randint(0, 20, 5)
rand_b = sorted(rand_a)

zipped = list(zip(names, rand_a, rand_b))
print("zipped:", zipped)

# dict from zip
hash_example = dict(zip(names, rand_a))
print("hash_example:", hash_example)

# Loop over zipped data (tuple unpacking)
for name, v1, v2 in zipped:
    print(f"Name {name}: ({v1}, {v2})")

# np.array vs np.asarray
npnames = np.array(names)
print("npnames:", npnames, "dtype:", npnames.dtype, "shape:", npnames.shape)

arr = np.arange(5)
copy_arr = np.array(arr)     # copy
view_arr = np.asarray(arr)   # no copy for ndarray

copy_arr[0] = 999
print("original arr:", arr)      # unchanged
print("copy_arr:", copy_arr)
```

## Functions

```python
import numpy as np

def Kelvin_2_Celsius(T):
    """Convert Kelvin to Celsius."""
    return T - 273.15

Temp = 273.16
print(f"{Temp} K are {Kelvin_2_Celsius(Temp):.2f} °C")


def Fahren_2_Kelvin(Temp):
    """
    Convert Fahrenheit to Kelvin.

    Input:
        Temp: temperature in degrees Fahrenheit
    """
    return ((Temp - 32.0) * (5.0 / 9.0)) + 273.15

print("Freezing point:", Fahren_2_Kelvin(32), "K")
print("Boiling point:", Fahren_2_Kelvin(212), "K")
print("Freezing point in °C:", Kelvin_2_Celsius(Fahren_2_Kelvin(32)))


def bmi_range(weight, height):
    """
    Compute BMI category from weight (kg) and height (m).
    """
    def bmi_val(w, h):
        return w / h**2

    bmi_value = bmi_val(weight, height)

    if bmi_value < 15:
        return "Very severely underweight"
    elif bmi_value < 16:
        return "Severely underweight"
    elif bmi_value < 18.5:
        return "Underweight"
    elif bmi_value < 25:
        return "Normal (healthy weight)"
    elif bmi_value < 30:
        return "Overweight"
    elif bmi_value < 35:
        return "Obese Class I"
    elif bmi_value < 40:
        return "Obese Class II"
    else:
        return "Obese Class III"

print("BMI category:", bmi_range(70, 1.80))
print("Docstring of Fahren_2_Kelvin:\n", Fahren_2_Kelvin.__doc__)
```

## Positional & keyword args

```python
import numpy as np

def Magnus(Temp, ice=False):
    """
    Saturation vapor pressure E(T) in hPa using Magnus formula.
    Temp in Celsius.
    """
    (A, B, C) = (22.587, 273.86, 6.1121) if ice else (17.625, 243.04, 6.1094)
    return C * np.exp((A * Temp) / (B + Temp))

print(Magnus(35))
print(Magnus(35, ice=True))
print(Magnus(Temp=35, ice=False))

# ---- BAD: mutable defaults ----
def ftest_bad(keyw_arg_0=[], keyw_arg_1=["2222"]):
    keyw_arg_0.append("0000")
    keyw_arg_1.append("1111")
    return keyw_arg_0, keyw_arg_1

print("bad defaults:", ftest_bad.__defaults__)
print(ftest_bad())
print(ftest_bad())  # grows again (side effect!)

# ---- GOOD: None defaults ----
def ftest_good(keyw_arg_0=None, keyw_arg_1=None):
    if keyw_arg_0 is None:
        keyw_arg_0 = []
    if keyw_arg_1 is None:
        keyw_arg_1 = ["2222"]
    keyw_arg_0.append("0000")
    keyw_arg_1.append("1111")
    return keyw_arg_0, keyw_arg_1

print("good defaults:", ftest_good.__defaults__)
print(ftest_good())
print(ftest_good())  # fresh each time

# *args example
def argeo_mean(first_value, *values):
    """Return (arithmetic_mean, geometric_mean) of the inputs."""
    amean = first_value
    gmean = first_value
    n = 1
    for v in values:
        amean += v
        gmean *= v
        n += 1
    return amean / n, gmean ** (1 / n)

print(argeo_mean(1))
print(argeo_mean(1, 2, 4))

args = [1, 2, 4, 32]
print(argeo_mean(*args))  # list unpacked into arguments

# **kwargs example
def f(a, b=0, **kwargs):
    print("kwargs:", kwargs)
    return a + b

print(f(1))
print(f(1, b=2))
print(f(1, b=2, c=34, d="test", e=None))
```

## Returning multiple values

```python
def E_T_WI(Temp):
    """Return (E_over_water, E_over_ice)"""
    return Magnus(Temp, ice=False), Magnus(Temp, ice=True)

E_water, E_ice = E_T_WI(22)
print("tuple return unpacked:", E_water, E_ice)

E_wice = E_T_WI(22)
print("tuple return:", E_wice)

def E_T_WI_hash(Temp):
    """Return dict with both values."""
    return {"E_Water": Magnus(Temp, ice=False), "E_Ice": Magnus(Temp, ice=True)}

E_dict = E_T_WI_hash(22)
print("dict return:", E_dict)
print("E_Water:", E_dict["E_Water"])
print("E_Ice:", E_dict["E_Ice"])
```

## Variable scope

```python
# Global variables
s = 10
t = 20
print("0:", s, t)

def function_t():
    a = 5  # local
    print("1:", a, s, t)
    return a

function_t()
print("2:", s, t)

# Local variable named 's' shadows global 's'
s = 10
t = 20
print("0:", s, t)

def function_t2():
    s = 5  # local shadow
    print("1:", s, t)
    return s

function_t2()
print("2:", s, t)

# Nested function: inner can read outer variables (closure), but not vice versa
a = 10
b = 20

def outer():
    c = 5
    print("outer:", a, b, c)

    def inner():
        local_a = 77
        print("inner:", local_a, b, c)

    inner()
    return c

print("outer returns:", outer())

# global keyword (use carefully)
s = 10
print("before:", s)

def change_global_s():
    global s
    s = 5

change_global_s()
print("after:", s)
```

## Function annotations

```python
def Magnus_annot(Temp: "°C", ice: "E over ice (T) or water (F)" = False) -> "hPa":
    (A, B, C) = (22.587, 273.86, 6.1121) if ice else (17.625, 243.04, 6.1094)
    return C * np.exp((A * Temp) / (B + Temp))

print("annotations:", Magnus_annot.__annotations__)

T_val = 10
print("E({} {}) = {:.3f} {}".format(
    T_val,
    Magnus_annot.__annotations__["Temp"],
    Magnus_annot(T_val),
    Magnus_annot.__annotations__["return"]
))
```

## Functions are references

```python
# Renaming a function: new reference points to same function
bmi_result = bmi_range
print("bmi_result:", bmi_result(45, 1.55))

def call_function(f, temp):
    print("Calling function", f.__name__, "on temp", temp)
    return f(temp)

print(call_function(Fahren_2_Kelvin, 44))
print(call_function(Kelvin_2_Celsius, 44))

def apply_trig(trig_func, exponent, angle):
    return trig_func(angle) ** exponent

angles = np.linspace(0, 2*np.pi, 10)
print(apply_trig(np.sin, 2, angles))
print(apply_trig(np.cos, 2, angles))

# Function returning a function (closure)
def f_0(a_value):
    def f_1(x):
        return a_value * x * (-a_value + x)  # a*x*(x-a)
    return f_1

g1 = f_0(1)
g2 = f_0(2)
print(g1(20), g2(10))
```

## ellipse + Taylor example

```python
import matplotlib.pyplot as plt
import numpy as np

def ellipse(a_value, b_value):
    def f_ell(x):
        return b_value * np.sqrt(1 - (x / a_value)**2)
    return f_ell

cal_ell = ellipse(2, 1)
x_val = np.linspace(-2, 2, 220)
upper = cal_ell(x_val)

plt.plot(x_val, upper)
plt.plot(x_val, -upper)
plt.axis("equal")
plt.title("Ellipse from function closure")
plt.show()


def taylor_f(x0, *coef_values):
    def f_t(x):
        res = 0
        for index, coef in enumerate(coef_values):
            res += coef * (x - x0)**index
        return res
    return f_t

sin_0 = taylor_f(0, 0, 1)
sin_1 = taylor_f(0, 0, 1, 0, -1/6)
sin_2 = taylor_f(0, 0, 1, 0, -1/6, 0, 1/120)

x_val = np.linspace(0, np.pi, 220)
plt.plot(x_val, np.sin(x_val), label="sin(x)")
plt.plot(x_val, sin_0(x_val), label="Order 1")
plt.plot(x_val, sin_1(x_val), label="Order 3")
plt.plot(x_val, sin_2(x_val), label="Order 5")
plt.legend()
plt.title("Taylor approximations to sin(x)")
plt.show()
```
