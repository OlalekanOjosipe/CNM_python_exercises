# Exercise 1 (Year data)

```python
year_data = metdata_orig[:, 0]
year_data
```

# Exercise 2 (min, max, and avg Mar temp)

```python
march_data = metdata_orig[:, 3]
```

```python
march_min = np.min(march_data)
```

```python
march_max = np.max(march_data)
```

```python
march_mean = np.mean(march_data)
```

```python
print("March minimum temperature:", march_min, "\nMarch maximum temperature:", march_max, "\nMarch mean temperature:", march_mean)
```

# Exercise 3 (average monthly and annual temperatures)

```python
temp_data = metdata_orig[:, 1:]

monthly_mean = temp_data.mean(axis=0)  # shape (12)
yearly_mean = temp_data.mean(axis=1)   # shape (136)
```

```python
print("Monthly mean:", monthly_mean)
```

```python
print("Yearly mean:", yearly_mean)
```

# Exercise 4 (Max and Min temp for 2000)

```python
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

year_2000 = np.where(year_data == 2000)[0][0]   # row index for year 2000
temp_2000 = temp_data[year_2000, :]

imax = np.argmax(temp_2000)
imin = np.argmin(temp_2000)

print("Year:", int(year_data[year_2000]))
print("Hottest month:", months[imax], f"({temp_2000[imax]:.2f})")
print("Coldest month:", months[imin], f"({temp_2000[imin]:.2f})")
```
