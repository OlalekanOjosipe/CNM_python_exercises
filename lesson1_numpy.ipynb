# Python Lesson 1 – NumPy Exercises


```python
import numpy as np
```

```python
metdata_orig = np.loadtxt("T_Alicante_EM.csv", delimiter=",", skiprows=1)

metdata_orig
```

```python
year_data = metdata_orig[:, 0]
temp_data = metdata_orig[:, 1:]

print(year_data[:5])
print(temp_data.shape)
```

## Exercise 1.1 – Monthly statistics for the year 1961

```python
temp_data[0, :]

print("1961 maximum temperature:", temp_data[0, :].max())
print("1961 minimum temperature:", temp_data[0, :].min())
print(f"1961 mean temperature: {temp_data[0, :].mean():.2f}")
```

```python
temp_data[0, :]

print('maximum monthly temperature:', np.max(temp_data[0, :]))
print('minimum monthly temperature:', np.min(temp_data[0, :]))
print(f"average monthly temperature: {np.mean(temp_data[0, :]):.2f}")
```

## Exercise 1.2 – March statistics across all years

```python
march_data = metdata_orig[:, 3]

print("March mean temperature:", march_data.mean())
print("March maximum temperature:", march_data.max())
print("March minimum temperature:", march_data.min())
```

## Exercise 1.3 – Yearly and monthly averages

```python
yearly_mean = temp_data.mean(axis=1)   # shape (136,)
monthly_mean = temp_data.mean(axis=0)  # shape (12,)

print("Yearly mean shape:", yearly_mean.shape)
print("Monthly mean shape:", monthly_mean.shape)

print("\nFirst 20 yearly means:")
print(yearly_mean[:20])

months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
print("\nMonthly means:")
for m, val in zip(months, monthly_mean):
    print(f"{m}: {val:.2f}")
```

## Exercise 1.4 – Hottest and coldest month in 2000

```python
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

idx_2000 = np.where(year_data == 2000)[0][0]   # row index for year 2000
temps_2000 = temp_data[idx_2000, :]

imax = np.argmax(temps_2000)
imin = np.argmin(temps_2000)

print("Year:", int(year_data[idx_2000]))
print("Hottest month:", months[imax], f"({temps_2000[imax]:.2f})")
print("Coldest month:", months[imin], f"({temps_2000[imin]:.2f})")
```

```python
yearly_mean = temp_data.mean(axis=1)

imax_year = np.argmax(yearly_mean)
print("Hottest year overall:", int(year_data[imax_year]))
print("Average temperature:", f"{yearly_mean[imax_year]:.2f}")
```
