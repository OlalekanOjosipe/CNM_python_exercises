```python
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image

print("Libraries imported successfully")
```

## 1. Working with images

```python
imgarray = np.array(Image.open("iberian-lynx.png"))

plt.imshow(imgarray)
plt.axis("off")   # hides axes
```

```python
imgarray.shape
```

```python
red = imgarray[:, :, 0]
green = imgarray[:, :, 1]
blue = imgarray[:, :, 2]

plt.imshow(red, cmap="gray")
plt.title("Red channel")
plt.colorbar()
plt.show()

plt.imshow(green, cmap="gray")
plt.title("Green channel")
plt.colorbar()
plt.show()

plt.imshow(blue, cmap="gray")
plt.title("Blue channel")
plt.colorbar()
plt.show()
```

```python
print("Red mean:", red.mean())
print("Green mean:", green.mean())
print("Blue mean:", blue.mean())
```

```python
print("Green max:", green.max())
print("Red max:", red.max())
print("Blue max:", blue.max())
```

## 2. More about NumPy

```python
a = np.ones((3,3))
b = np.zeros((2,4))
c = np.full((3,2), 7)
d = np.eye(4)

print(a)
print()
print(b)
print()
print(c)
print()
print(d)
```

```python
x = np.arange(0, 10, 2)
y = np.linspace(0, 10, 5)

print("arange:", x)
print("linspace:", y)
```

```python
A = np.random.normal(loc=2, scale=1, size=(3,3))
B = np.random.normal(loc=2, scale=1.5, size=(3,3))

print("A:\n", A)
print("\nB:\n", B)

C = np.sqrt(A**2 + B**2)

print("\nResult:\n", C)
```

```python
sub = A[:2, :2]
sub_copy = A[:2, :2].copy()

sub[:] = 100
print("A after modifying sub:\n", A)

sub_copy[:] = 0
print("\nA after modifying sub_copy:\n", A)
```

```python
M = np.arange(100).reshape(10,10)
print(M)

F = M[2:5, [9,5,2]]
print("\nFancy indexed:\n", F)
```

```python
print(A)
print("\nTranspose:\n", A.T)

print("\nDot product:\n", np.dot(A.T, A))
print("\nMatmul:\n", A.T @ A)
```

```python
X = np.random.randn(2,4)
Y = np.random.randn(2,4)

print("Horizontal:\n", np.hstack((X,Y)))
print("\nVertical:\n", np.vstack((X,Y)))
```

```python
Z = np.random.randn(4,4)

r = Z.ravel()
f = Z.flatten()

r[0] = 999
f[2] = 45

print("Z after ravel:\n", Z)
print("f after ravel:\n", f)
print("Z after ravel:\n", Z)
```

```python
mask = Z > 0
print(mask)

Z2 = Z.copy()
Z2[mask] = 1
Z2[~mask] = 0

print("\nBinary map:\n", Z2)
```

```python
metdata_orig = np.loadtxt("T_Alicante_EM.csv", delimiter=",", skiprows=1)
metdata = metdata_orig[:, 1:]   # remove year column

metdata.shape
```

## Exercise 2.1 â€“ Heatmap of monthly temperatures

```python
plt.imshow(metdata, cmap="BrBG", aspect="auto")
plt.colorbar(label="Temperature (Â°C)")
plt.title("Monthly temperature heatmap")
plt.xlabel("Month")
plt.ylabel("Year index")
plt.show()
```

## Exercise 2.2 â€“ Monthly and annual averages

```python
ave_monthly = np.mean(metdata, axis=0)

plt.plot(ave_monthly, marker='o')
plt.title("Average monthly temperature")
plt.xlabel("Month (0 = Jan)")
plt.ylabel("Temperature (Â°C)")
plt.grid(True)
plt.show()
```

```python
ave_annual = np.mean(metdata, axis=1)

plt.plot(ave_annual)
plt.title("Average annual temperature")
plt.xlabel("Year index")
plt.ylabel("Temperature (Â°C)")
plt.grid(True)
plt.show()
```

## Exercise 2.3 â€“ Monthly max and min

```python
max_monthly = np.max(metdata, axis=0)
min_monthly = np.min(metdata, axis=0)

plt.plot(max_monthly, label="Max monthly")
plt.plot(min_monthly, label="Min monthly")
plt.title("Monthly max and min temperatures")
plt.xlabel("Month")
plt.ylabel("Temperature (Â°C)")
plt.legend()
plt.grid(True)
plt.show()
```

## Exercise 2.4 â€“ Annual max and min

```python
max_annual = np.max(metdata, axis=1)
min_annual = np.min(metdata, axis=1)

plt.plot(max_annual, label="Max annual")
plt.plot(min_annual, label="Min annual")
plt.title("Annual max and min temperatures")
plt.xlabel("Year index")
plt.ylabel("Temperature (Â°C)")
plt.legend()
plt.grid(True)
plt.show()
```

## Exercise 2.5 â€“ Multi-panel summary figure

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))
fig.tight_layout(pad=3.0)

# Top-left: max monthly
ax[0, 0].plot(max_monthly)
ax[0, 0].set_title("Max monthly")
ax[0, 0].set_xlabel("Month (0-11)")
ax[0, 0].set_ylabel("Temp (Â°C)")
ax[0, 0].grid(True)

# Top-right: min monthly
ax[0, 1].plot(min_monthly)
ax[0, 1].set_title("Min monthly")
ax[0, 1].set_xlabel("Month (0-11)")
ax[0, 1].set_ylabel("Temp (Â°C)")
ax[0, 1].grid(True)

# Bottom-left: max annual
ax[1, 0].plot(max_annual)
ax[1, 0].set_title("Max annual")
ax[1, 0].set_xlabel("Year index")
ax[1, 0].set_ylabel("Temp (Â°C)")
ax[1, 0].grid(True)

# Bottom-right: min annual
ax[1, 1].plot(min_annual)
ax[1, 1].set_title("Min annual")
ax[1, 1].set_xlabel("Year index")
ax[1, 1].set_ylabel("Temp (Â°C)")
ax[1, 1].grid(True)

fig.suptitle("Alicante Temperature Dataset Summary", y=1.02)
plt.show()
```
